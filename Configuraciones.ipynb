Configuración Lineal

```python
import numpy as np
import random as rd
import pandas as pd

def rad_to_angle(a):
    return a * 180 / np.pi

def find_theta(a: float, b: float): #(-1,0)
    if (b == 0): theta = np.arccos(a/abs(a))
    elif (a == 0): theta = np.arcsin(b / abs(b))
    else:
        theta = np.arctan(b / a)
        if (a < 0):
            theta += np.pi
    return theta

def vec_r(A: list, B: list):
    x_v = B[0] - A[0] ; y_v = B[1] - A[1]
    theta = 0 if (x_v == 0 and y_v == 0 ) else find_theta(x_v, y_v)
    norma = x_v**2 + y_v**2
    return norma, theta

def q_interaction(q: int, q0: list):
    q_i = [[10, 10], [19, 10]] ; t = m = xi = yi = 0 ; ke = 8.9875e9 ; q0 = list(q0)
    for i in range (2):
        if (q0 in q_i):
            return ["udf", "udf"]
        m, t = vec_r(q_i[i], q0)
        xi += (-1.0)**i * np.cos(t) / m # m ya esta al cuadrado
        yi += (-1.0)**i * np.sin(t) / m # El -1 contempla la existencia de cargas negativas
    t = 0 if (xi == 0 or yi == 0 ) else find_theta(xi, yi)
    m = abs(ke * q) * (xi**2 + yi**2)**0.5 ; t = rad_to_angle(t)
    E = format(m, ".2e") ; t = round(t)
    return [E, t]

def points(a: int, b: int, c: int):
    puntos = set()
    while len(puntos) <  c:
        punto = (rd.randint(8, 21), rd.randint(8, 12))
        if punto not in puntos:
            puntos.add(punto)
    puntos = sorted(puntos, key=lambda p: (p[0]**2 + p[1]**2)**0.5)
    return puntos

q_charge = 1e-6 #Coulombs

Puntos = points(-4, 12, 50)
Resultados = []
C = ["Magnitud", "Angulo"]
for i in Puntos:
    Resultados.append(q_interaction(q_charge, i))
lC = pd.DataFrame(Resultados, index = Puntos, columns = C)
lC
```
